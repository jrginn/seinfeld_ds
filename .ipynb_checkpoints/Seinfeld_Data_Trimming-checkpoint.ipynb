```python
import pandas as pd
import numpy as np
```

```python
# load csv
df = pd.read_csv('full_lines.csv', index_col=0)
```

```python
df
```

```python
full_names = df.character.unique().astype(str)
for n in full_names:
    if n.startswith('J') or n.startswith('G') or n.startswith('E') or n.startswith('K'):
        print(n)
    
```

```python
j_names = ['JERY', 'JERR']
g_names = ['GEOGE', "GOERGE", 'GEROGE']
e_names = ["ELANE", "ELAIEN", 'EALINE', 'ELIANE']
k_names = ['KESSLER', 'KRMAER']
names = j_names + g_names + e_names + k_names
typos = {
    'JERRY': j_names,
    'GEORGE': g_names,
    'ELAINE': e_names,
    'KRAMER': k_names
}
```

```python
for t in typos:
    print('KEY: ' + str(t))
    for x in typos[t]:
        print(x)
```

```python
# df.loc[df['character']=='JERY', 'character'] = 'JERRY'
for t in typos:
    for x in typos[t]:
        df.loc[df['character']==x, 'character'] = t
```

```python
hope = df.character.unique()
for y in names:
    if y in hope:
        print('DRAT')
```

```python
df = df.loc[(df['character']=='JERRY') | (df['character']=='GEORGE') | (df['character']=='ELAINE') | (df['character']=='KRAMER')]
df
```

```python
df = df.reset_index()
df
```

```python
df = df.drop(columns=['index'])
df
```

```python
df.to_csv('jgek_lines.csv')
```
