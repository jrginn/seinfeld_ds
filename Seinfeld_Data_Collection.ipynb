```python
import os
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Get script html files into a list of files
path = './scripts/'
files = []
# list storing lists (episodes) of lists (character, line, episode number)
all_data = []
# loops through scripts directory and uses each html file to extract dialogue
for f in os.listdir(path):
    if f.endswith('.shtml'):
        # takes episode number and handles double episodes
        epi_num = f.split('-')[1].split('.')[0]
        if 'and' in epi_num:
            epi_num = epi_num.split('and')[0]
        # creates file object and adds it to list
        new_file = open(path + f, 'r+')
        files.append(new_file)
        # uses BeautifulSoup to parse through html
        soup = BeautifulSoup(new_file, "html.parser")
#         words = re.split('=+',soup.get_text())[1].split('\n\n\t\t\t\t\t\t\t\n\t\t\t\t\t\t')
        # gets all text removed of formatting from seinology
        words = re.split("(?:\\n){2,}(?:\\t){2,}\\n(?:\\t){2,}", re.split('=+',soup.get_text())[1])
        # flag to include opening monologues not labelled as Jerry
        jerrynext = False
        # list to store data for each episode
        epi_data = []
        # loops through each line of the text and extracts characters' lines
        for l in words:
            # list to store data about each line
            line_data = []
            # removes words in parentheses and leftover newline characters
            l = re.sub("\([^)]*\)", '', l)
            l = re.sub('\\n', '', l)
            # checks if the line following this one will be part of Jerry's standup
            if l == '[Opening Monologue]':
                jerrynext = True
            elif jerrynext:
                line_data.append('JERRY')
                line_data.append(l)
                line_data.append(int(epi_num))
                jerrynext = False
            elif re.match(r'(?:[A-Z]+\.?\s?)*[A-Z]*:(?:\s*:)*', l):
                # this regex checks for all caps words followed by one or more colons with potential spaces between them
                if '&' in re.split('\s*:', l)[0] or ' AND ' in  re.split('\s*:', l)[0]:
                    # splits on & or AND adds each character with their shared line
                    epi_data.append([re.split('\s*(?:&|AND)\s*', re.split('\s*:', l)[0])[0], re.split('\s*:', l)[1], int(epi_num)])
                    epi_data.append([re.split('\s*(?:&|AND)\s*', re.split('\s*:', l)[0])[1], re.split('\s*:', l)[1], int(epi_num)])
                else:
                    # adds character name
                    line_data.append(re.split(':(?:\s*:)*', l)[0])
                    # adds line
                    line_data.append(re.split(':(?:\s*:)*', l)[1])
                    #adds episode number
                    line_data.append(int(epi_num))
            if len(line_data) > 0:
                # ensures that only character lines are added
                epi_data.append(line_data)
        # appends the list of lines to all data
        all_data.append(epi_data)
        
```

```python
# put every episodes' lines in one list to help create the dataframe
lines = []
for d in all_data:
    for e in d:
        lines.append(e)
len(lines)
```

```python
# creates dataframe
cols = ['character', 'line', 'episode number']
frame = pd.DataFrame(lines, columns=cols)
frame
```
